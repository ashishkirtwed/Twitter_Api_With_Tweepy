# Searching for a tweet

```python
query="#DataScience -@HaroldSinnott" #here - means to remove that content from the search 
for i,status in enumerate(tweepy.Cursor(api.search,q=query).items(50)):
    print(i,status.text,status.author.screen_name)
```

```python
query_username="DataScience" #here - means to remove that content from the search 
for i,user in enumerate(tweepy.Cursor(api.search_users,q=query,per_page=20).items(50)):
    print(i,user.screen_name)
```

```python
query_username="TheRock" #here - means to remove that content from the search 
def process_page(page_result):
    for i,user in enumerate(page_result):
        print(i,user.screen_name)
for i,page in enumerate(tweepy.Cursor(api.search_users,q=query_username,per_page=10).pages(3)):
    print(i,"page")
    process_page(page)
```

```python
import time
def limit_handled(cursor):
    while True:
        try:
            yield cursor.next()
        except tweepy.RateLimitError:
            time.sleep(15*60)
            
```

```python
for follower in limit_handled(tweepy.Cursor(api.followers).items()):
    print(follower.screen_name)
```
